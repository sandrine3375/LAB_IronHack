#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
from matplotlib.pyplot import figure

f, [ax1, ax2] = plt.subplots(1,2)  #pour avoir deux tableaux differents

f.set_figheight(5)
f.set_figwidth(5)

ax1.plot(x,x*2)
ax1.set_title('LINEAR PLOT')   
ax1.set_xlabel('x label')     
ax1.set_ylabel('y label')   


ax2.plot(x,np.exp(x))
ax2.set_title('linear PLOT')
ax2.set_xlabel('x label')
ax2.set_ylabel('z label')
```

#### As a bonus challenge, try to add a legend to the plot.

```python
from matplotlib.pyplot import figure

f, [ax1, ax2] = plt.subplots(1,2)  #pour avoir deux tableaux differents

f.set_figheight(5)
f.set_figwidth(5)

ax1.plot(x,x*2,label='LEG') #POUR LA LEGENDE LABEL LEG
ax1.set_title('LINEAR PLOT')
ax1.legend()
ax1.set_xlabel('x label')     
ax1.set_ylabel('y label')   


ax2.plot(x,np.exp(x), label='LEG2')
ax2.legend()
ax2.set_title('linear PLOT')
ax2.set_xlabel('x label')
ax2.set_ylabel('z label')
```
